```python
df_tillnow1['rating'] = df_tillnow1['rating'].astype(float)
```

```python
movie_pivot = df_tillnow1.pivot_table(columns='userId', index = 'title', values = 'rating')
```

```python
movie_pivot.fillna(0, inplace = True)
```

```python
from scipy.sparse import csr_matrix
```

```python
movie_sparse = csr_matrix(movie_pivot)
```

```python
movie_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(n_neighbors = 7, algorithm = 'brute', metric = 'cosine')
```

```python
model.fit(movie_sparse)
```

```python
df_tillnow1.drop(columns = ['genres', 'userId', 'rating'], inplace = True)
```

```python
df_tillnow1.drop_duplicates(inplace = True)
```

```python
df1 = df_tillnow1.copy()
ti = list()

for i in df1['title']:
    ti.append(i.split(' ()')[0])


df1['title'] = ti
```

```python
df1
```

```python
for i in range(len(suggestions)):
    print(movie_pivot.index[suggestions[i]])
```

```python
def recommendation(movie_name):
    movie_id = df1[df1['title']==movie_name].drop_duplicates('title')['movieId'].values[0]
    distances, suggestions = model.kneighbors(movie_pivot.iloc[movie_id, :].values.reshape(1, -1))
    
    
    for i in range(len(suggestions)):
        return (movie_pivot.index[suggestions[i]])
```

```python
res = recommendation('Jumanji (1995)')
```
