# General Exploration

```python
def basic_info(data):
    print("Data shape is: ", data.shape)
    print("Data size is: ", data.size)
    print(data.info())
    print("Number of null values are: ", data.isnull().sum())
    cat, num = list(), list()
    for i in data.columns:
        if data[i].dtype == object:
            cat.append(i)
        else:
            num.append(i)
    print(cat, "\n")
    print(num)
    return cat, num
```

```python
categorical, numerical = basic_info(iris)
```

```python
df_iris = iris.copy(deep = True)
```

```python
df_iris['Species'].value_counts().index.tolist() # getting the types of species
```

```python
replacing = {"Species": {'Iris-setosa': 1, 'Iris-virginica': 2, 'Iris-versicolor': 3}} #doing integer-hot encoding
```

```python
df_iris.replace(replacing, inplace= True) #replacing categorical values with repective integer value
```

```python
df_iris
```

```python
df_iris = df_iris.drop(['Id'], axis = 1) # dropping Id as it will not be required for model
```

# Splitting

```python
from sklearn.model_selection import train_test_split # for splitting the data into training and testing
```

```python
X = df_iris.iloc[:, :4]
```

```python
X
```

```python
y = df_iris.iloc[:, -1]
```

```python
y
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.20, random_state=42) # taking 80% as training and rest testing
```

```python
test_x
```

```python
test_y
```

# Model Making

```python
from sklearn.naive_bayes import GaussianNB #importing naive_bayes
```

```python
gnb = GaussianNB()
```

```python
model_gnb = gnb.fit(train_x, train_y) #fitting the model
```

```python
y_gnb_predict = model_gnb.predict(test_x) # predicitng the output
```

```python
GNB_df = pd.DataFrame(data = {"Actual":np.array(test_y), "Predicted": y_gnb_predict}) # making a df for actual VS prediction
```

```python
GNB_df
```

```python
model_gnb.score(test_x, test_y) # Checking the score
```
