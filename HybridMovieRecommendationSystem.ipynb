```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
def recommendation_part1_CF(movie_list, movie_ratings):
    #making dictionary from movie list and rating
    dic_watched1 = {}
    for key in movie_list:
        for value in movie_rating:
            dic_watched1[key] = value
            movie_rating.remove(value)
            break
    
    #Making a copy of original movie similarity matrix, so that the original stays intact
    #Keeping deep = True, so that a separate copy is made
    movie_similarity_matrix_cf = movie_similarity_matrix.copy(deep = True)
    #print(movie_similarity_matrix_cf)
    
    #Now adding our dictionary as an row
    movie_similarity_matrix_cf = movie_similarity_matrix_cf.append(pd.DataFrame(dic_watched1), ignore_index=True)
    #print("\n\nBefore Nan\n\n\n",movie_similarity_matrix_cf)
    #print("\n\n", movie_similarity_matrix_cf.isnull().sum())
    
    #making nan to 0.0
    movie_similarity_matrix_cf.fillna(0, inplace = True)
    #print("\n\n\nAfter Nan\n\n\n")
    #print(movie_similarity_matrix_cf)
    
    #Creating sparse matrix on which our NearestNeighbors model will get fit
    movie_sparse_cf = csr_matrix(movie_similarity_matrix_cf)
    
    #Creating our NearestNeighbors model
    modelr = NearestNeighbors(n_neighbors = 11, algorithm = 'brute', metric = 'euclidean')
    #Fitting our model
    modelr.fit(movie_sparse_cf)
    
    #lets get the similar users
    distance, similar_users = modelr.kneighbors(movie_similarity_matrix_cf.iloc[-1].values.reshape(1, -1))
    
    #print(distance, "\n", similar_users)
    
    returning_similar_users = list()
    for i in range(1,len(similar_users[0])):
        returning_similar_users.append(similar_users[0][i])
    #print(returning_similar_users)
    
    similar_users_df = movie_similarity_matrix_cf.iloc[similar_users[0][1]].to_frame()
    similar_users_df.columns = ['ratings']
    for i in range(2, len(similar_users)):
        subdf = movie_similarity_matrix_cf.iloc[similar_users[0][i]].to_frame()
        subdf.columns = ['ratings']
        similar_users_df = similar_users_df.add(subdf, fill_value = 0)
        
    similar_users_df = similar_users_df[similar_users_df['ratings'] > 1]
    #print("\n\n\n", similar_users_df.head(15))
    
    
    #getting movies watched by similar users (contain names of movies watched by our user) 
    list_of_movies_by_similar_users = similar_users_df.index.tolist()
    #print(list_of_movies_by_similar_users, "\n\n\n")
    #list_of_movies_ratings_ = similar_users_df['ratings'].tolist()
    
    #remvoing already watched movies
    list_of_movies_by_similar_users = list(set(list_of_movies_by_similar_users) - set(movie_list))
    list_of_movies_ratings_ = similar_users_df.loc[list_of_movies_by_similar_users, 'ratings'].tolist();
    #print(len(list_of_movies_by_similar_users), "\nw", len(list_of_movies_ratings_))
    
    recommendation_part2_CB_and_Recommendation(movie_list, movie_ratings, list_of_movies_by_similar_users, list_of_movies_ratings_)
```

```python
def recommendation_part2_CB_and_Recommendation(movie_list, movie_ratings, list_of_movies_by_similar_users, list_of_movies_ratings_):
    datauser = content_based_df_main.loc[content_based_df_main['title'].isin(movie_list)] #already watched movie
    datarecommendation = content_based_df_main.loc[content_based_df_main['title'].isin(set(list_of_movies_by_similar_users))] #from where the recommendation will come
    #print("\n\n\nDataUser\n\n")
    #print(datauser.shape)
    #print("\n\n\nDataRecommendation\n\n")
    #print(datarecommendation.shape)
    #print("\n\n", datauser.isnull().sum())  #no null
    #print("\n\n", datarecommendation.isnull().sum()) #no null
    
    #Replacing the rating in datauser (as its an average rating prior) with the movie_ratings we took as an arguement
    #print(len(list_of_movies_by_similar_users), " " , len(datarecommendation['title'].tolist()))
    datauser.loc[:,'rating'] = movie_ratings
    #datarecommendation.loc[:, 'rating'] = list_of_movies_ratings_
    #print("\n\n\nDataUser After Rating Updation\n\n")
    #print(datauser)
    
    #print(list_of_movies_by_similar_users, "\n\n")
    #print(datarecommendation['title'].tolist())
    #print(len(list_of_movies_by_similar_users), " " , len(datarecommendation['title'].tolist()))
    
    #Now taking out rating specifically for multiplication purpose
    list_user_rating = np.array(datauser['rating'])
    #print("\n\n", np.isnan(np.sum(list_user_rating))) #no null
    
    #making weighted matrix for datauser which will be multiplied by list_user_rating
    datauser_genre_matrix = datauser.drop(['movieId', 'title', 'rating'], axis=1)
    #print(datauser_genre_matrix)
    #print("\n\n", datauser_genre_matrix.isnull().sum()) #no nan
    
    list_user_rating = list_user_rating.reshape(len(list_user_rating), 1) #reshaping the matrix so that multiplication can happen
    #print(list_user_rating)
    
    #changing our datauser_genre_matrix into numpy array so that multiplication is possible
    data_user_genre_matrix_numpy = np.array(datauser_genre_matrix)
    #print(data_user_genre_matrix_numpy)
    
    #now multiplying our rating with our numpy matrix (also called weighted array)
    data_user_multiply = np.multiply(list_user_rating, data_user_genre_matrix_numpy)
    #print(data_user_multiply)
    
    #now making a weighted sumed matrix
    data_user_sum_weighted_matrix = np.sum(data_user_multiply, axis = 0) ##using np.sum() so as to get column wise sum
    #print(data_user_sum_weighted_matrix)
    
    #now making the ratings normalized
    normalized_data_user_sum_weighted_matrix = data_user_sum_weighted_matrix/sum(data_user_sum_weighted_matrix)
    #print(normalized_data_user_sum_weighted_matrix)
    
    """The previous step is done"""
    """Now by using normalized_data_user_sum_weighted_matrix, we will try to recommend the movies to user"""
    
    #making a new dataframe, where only genre scores are present
    datarecommedation_genre_frame = datarecommendation.drop(['movieId', 'title', 'rating'], axis=1)
    #print(datarecommedation_genre_frame)
    
    #making recommendation numpy array for multiplication
    data_recommendation_genre_matrix_numpy = np.array(datarecommedation_genre_frame)
    #print(data_recommendation_genre_matrix_numpy)
    
    #now making multiplied recommendation genre matrix
    data_recommendation_genre_matrix_multiply = np.multiply(normalized_data_user_sum_weighted_matrix, data_recommendation_genre_matrix_numpy)
    #print(data_recommendation_genre_matrix_multiply)
    
    #now making weighted recommendation matrix
    data_recommendation_weighted_matrix = np.sum(data_recommendation_genre_matrix_multiply, axis = 1)

    
    """Now since we got the recommendation matrix, now we will merge the matrix as a column in the datarecommendation dataframe"""
    datarecommendation['recommendation_rating'] = pd.Series(data_recommendation_weighted_matrix)
    #print(datarecommendation.head())
    datarecommendation = datarecommendation.sort_values('recommendation_rating', ascending = False)
    #print(datarecommendation.head())
    print(datarecommendation[['title', 'recommendation_rating']].head(10))
    #print(list(datarecommendation['title']))
    
```
