```python
def grouped_plot_of_no_of_atm_and_pos_wrt_year_and_month(year,data,lisofbank,month):
    # year= integer
    # data= pandas dataframe
    # lisofbank= list of string
    # month= string
    datanew=data.loc[(data["bank_name"].isin(lisofbank)) & (data["year"]==year) & (data["month"]==month)]
    datanew["no_of_atm"]=datanew["no_atms_on_site"]+datanew["no_atms_off_site"]
    datanew["no_of_pos"]=datanew["no_pos_on_line"]+datanew["no_pos_off_line"]
    
    plt.figure(figsize=(20,6))
    barwidth=0.25
    #fig,ax=plt.subplots()
    #bar1=datanew["no_of_atm"]
    #bar2=datanew["no_of_pos"]
    bar1=[]
    bar2=[]
    for i in lisofbank:
        subdata=datanew.loc[datanew["bank_name"]==i]
        bar1+=[subdata["no_of_atm"].sum()]
        bar2+=[subdata["no_of_pos"].sum()]

    r1=np.arange(len(bar1))
    r2=[x+barwidth for x in r1]

    plt.bar(r1, bar1, color="red", width=barwidth, edgecolor="yellow", label="ATM DEPLOYED")
    plt.bar(r2, bar2, color='cyan', width=barwidth, edgecolor='black', label="POS DEPLOYED")
    
    for index, value in enumerate(bar1):
        plt.text(index-0.13, value, str(value),color="black", fontweight="bold")
    
    for index, value in enumerate(bar2):
        plt.text(index+0.14,value, str(value),color="black", fontweight="bold")

    plt.title("Number of ATM and POS deployed for the month {0} and year {1}".format(month, year), fontweight="bold")
    plt.xlabel("Name of the banks")
    plt.xticks([r+barwidth for r in range(len(bar1))], lisofbank, rotation=-45)

    plt.legend()
    plt.show()
```

```python
#Now we will see year wise no. of credit and debit card issued
#Total number of credit cards issued outstanding (after adjusting the number of cards withdrawan/cancelled). #no_credit_cards
#Total number of debit cards issued outstanding (after adjusting the number of cards withdrawan/cancelled). #no_debit_cards
#main dataframe using dataS1
#Banks choosen for my research ["BANK OF BARODA", "PUNJAB NATIONAL BANK", "HDFC BANK LTD",'STATE BANK OF INDIA', 'ICICI BANK LTD']

import palettable.colorbrewer.qualitative as pcq #for the coloring of the donout chart

def no_of_debit_and_credit_card_wrt_year(year,data,lisofbanks, month):
    datanew=data.loc[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data['month']==month)]
    #print(datanew)
    plt.figure(figsize=(20,6))
    barwidth=0.25
    
    plt.figure(figsize=(5,5))
    my_circle=plt.Circle((0,0), .6, color="white")
    
    plt.pie([datanew["no_credit_cards"].sum(), datanew["no_debit_cards"].sum()], labels=["Number of credit cards {0}".format(datanew["no_credit_cards"].sum()),"Number of debit cards {0}".format(datanew["no_debit_cards"].sum())], autopct="%1.1f%%",  textprops={"fontsize":9,"fontweight":"bold"},colors=pcq.Paired_7.hex_colors)
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.title("No. of Credit & Debit cards issued by the banks in the year {0} of the month {1}".format(year, month))
    plt.legend()
    plt.show()
    
```

```python
no_of_debit_and_credit_card_wrt_year(2020,dataS1,["BANK OF BARODA", "PUNJAB NATIONAL BANK", "HDFC BANK LTD",'STATE BANK OF INDIA', 'ICICI BANK LTD'],"January")
```

```python
#So now we will select either credit or debit card. upon that choice these will be shown the total number of transaction

#if Credit:
#Total number of financial transactions done by the credit card issued by the bank at ATMs #no_credit_card_atm_txn
#Total number of financial transactions done by the credit card issued by the bank at POS terminals  #no_credit_card_pos_txn

#if Debit
#Total number of financial transactions done by the debit card issued by the bank at ATMs #no_debit_card_atm_txn
#Total number of financial transactions done by the debit card issued by the bank at POS terminals #no_debit_card_pos_txn

#Standard list of banks:
#Banks choosen for my research ["BANK OF BARODA", "PUNJAB NATIONAL BANK", "HDFC BANK LTD",'STATE BANK OF INDIA', 'ICICI BANK LTD']

def no_of_financial_transaction(year,data,lisofbanks,DorC, month):
    datanew=data.loc[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data["month"]==month)]
    bar1=[] #for total num of transaction
    if DorC=="credit" or DorC=="Credit":
        # since total num of transaction= total num of transaction in atm + total num of transaction in pos
        datanew["total_num_txn_credit"]=datanew["no_credit_card_atm_txn"]+datanew["no_credit_card_pos_txn"]
        
        #now combining all of the values of 5 banks
        for i in lisofbanks:
            subdata=datanew[datanew["bank_name"]==i]
            bar1+=[subdata["total_num_txn_credit"].sum()]
        print(bar1)
            
    else:
        # since total num of transaction= total num of transaction in atm + total num of transaction in pos
        datanew["total_num_txn_debit"]=datanew["no_debit_card_atm_txn"]+datanew["no_debit_card_pos_txn"]
        
        #now combining all of the values of 5 banks
        for i in lisofbanks:
            subdata=datanew[datanew["bank_name"]==i]
            bar1+=[subdata["total_num_txn_debit"].sum()]
            
        print(bar1)
            
    plt.figure(figsize=(11,8), facecolor='white') #background of the plot
    barwidth=0.25
    plt.gca().set_facecolor("cyan") #inside the plot
    r1=np.arange(len(bar1))
    #r2=[x+barwidth for x in r1]
    
    plt.bar(r1, bar1, color="blue", width=barwidth, edgecolor="yellow", label="Total Number of Transactions")
    #plt.bar(r2, bar2, color="green", width=barwidth, edgecolor="red", label="Total Value of Transaction")
    
    plt.title("Total Number of Transaction for the year {0} & month {2} for {1} cards".format(year,DorC,month), fontweight="bold")
    plt.xlabel("Bank Names")
    plt.ylabel("No. of Transaction")
    plt.xticks([r for r in range(len(bar1))],lisofbanks, rotation=-45)
    
    for index, value in enumerate(bar1):
        plt.text(index-0.13, value, str(value),color="black", fontweight="bold")
    
    
    plt.legend() #to show bar labels
    plt.show()
            
```

```python
import matplotlib as mlt

def val_of_financial_transaction(year,data,lisofbanks,DorC, month):
    datanew=data[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data["month"]==month)]
    bar1=[] #stroing th total value of transaction of each bank
    if DorC=="credit" or DorC=="Credit":
        #total value of transactions= total transaction value in atm (credit) + total transaction value in pos (credit)
        datanew["total_value_of_transaction_credit"]=datanew["no_credit_card_atm_txn_value_in_mn"]+datanew["no_credit_card_pos_txn_value_in_mn"]
        
        for i in lisofbanks:
            subdata=datanew.loc[datanew["bank_name"]==i]
            bar1+=[subdata["total_value_of_transaction_credit"].sum()]
            
    else:
        #total value of transctions= total transaction value in atm (debit) + total transaction value in pos (debit)
        datanew["total_value_of_transaction_debit"]=datanew["no_debit_card_atm_txn_value_in_mn"]+datanew["no_debit_card_pos_txn_value_in_mn"]
        
        for i in lisofbanks:
            subdata=datanew.loc[datanew["bank_name"]==i]
            bar1+=[subdata["total_value_of_transaction_debit"].sum()]
            
    
    plt.figure(figsize=(16,5))
    barwidth=0.25
    plt.gca().set_facecolor("#C1CDCD") #inside the plot
    r1=np.arange(len(bar1))
    print(r1)
    
    plt.bar(r1, bar1, color="#00008B", width=barwidth, edgecolor="#FFF8DC", label="Total Value of Transactions")
    #plt.axes(visible=True)
    
    plt.title("Total Value of Transaction for the year {0}  of the month {2} of {1} cards".format(year, DorC, month), fontweight="bold")
    plt.xlabel("Bank Names")
    plt.ylabel("Value of Transaction \nin Millions", visible=True)
    plt.xticks([r for r in range(len(bar1))],lisofbanks, rotation=-45)
    
    for index, value in enumerate(bar1):
        plt.text(index-0.13, value, str(value),color="black", fontweight="bold")
    
    
    plt.legend() #to show bar labels
    plt.show()
    
```

```python
def percentage_decrease_in_val_of_transaction(year,data,lisofbanks,DorC, month1, month2):
    datanew1=data[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data["month"]==month1)]
    datanew2=data[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data["month"]==month2)]
    bar1=[] #stroing th total value of transaction of each bank for month1
    bar2=[] #stroing th total value of transaction of each bank for month2
    if DorC=="credit" or DorC=="Credit":
        #total value of transactions= total transaction value in atm (credit) + total transaction value in pos (credit)
        datanew1["total_value_of_transaction_credit"]=datanew1["no_credit_card_atm_txn_value_in_mn"]+datanew1["no_credit_card_pos_txn_value_in_mn"]
        datanew2["total_value_of_transaction_credit"]=datanew2["no_credit_card_atm_txn_value_in_mn"]+datanew2["no_credit_card_pos_txn_value_in_mn"]
        
        for i in lisofbanks:
            subdata1=datanew1.loc[datanew1["bank_name"]==i]
            subdata2=datanew2.loc[datanew2["bank_name"]==i]
            bar1+=[subdata1["total_value_of_transaction_credit"].sum()]
            bar2+=[subdata2["total_value_of_transaction_credit"].sum()]
            
    else:
        #total value of transctions= total transaction value in atm (debit) + total transaction value in pos (debit)
        datanew1["total_value_of_transaction_debit"]=datanew1["no_debit_card_atm_txn_value_in_mn"]+datanew1["no_debit_card_pos_txn_value_in_mn"]
        datanew2["total_value_of_transaction_debit"]=datanew2["no_debit_card_atm_txn_value_in_mn"]+datanew2["no_debit_card_pos_txn_value_in_mn"]
        
        for i in lisofbanks:
            subdata1=datanew1.loc[datanew1["bank_name"]==i]
            bar1+=[subdata1["total_value_of_transaction_debit"].sum()]
            subdata2=datanew2.loc[datanew2["bank_name"]==i]
            bar2+=[subdata2["total_value_of_transaction_debit"].sum()]
            
    perbar=[]
    #print(bar1,bar2)
    zip_object=zip(bar1,bar2)
    #print(zip_object)
    for i, j in zip_object:
        perbar.append(abs(((j-i)/i)*100))
    return perbar, DorC
```

```python
valperdec, dorc=percentage_decrease_in_val_of_transaction(2020,dataS1,["BANK OF BARODA", "PUNJAB NATIONAL BANK", "HDFC BANK LTD",'STATE BANK OF INDIA', 'ICICI BANK LTD'],'credit',"January","February")
```

```python
def dec_over_coming_time_val(valperdec,DorC,numofiter,monthofstart,data,year,lisofbanks):
    datanew1=data[(data["bank_name"].isin(lisofbanks)) & (data["year"]==year) & (data["month"]==monthofstart)]
    bar1=[]
    if DorC=="credit" or DorC=="Credit":
        datanew1["total_value_of_transaction_credit"]=datanew1["no_credit_card_atm_txn_value_in_mn"]+datanew1["no_credit_card_pos_txn_value_in_mn"]
        for i in lisofbanks:
            subdata1=datanew1.loc[datanew1["bank_name"]==i]
            bar1+=[subdata1["total_value_of_transaction_credit"].sum()]
        for i in range(numofiter):
            zip_object=zip(bar1,valperdec)
            bar2=[]
            for j, k in zip_object:
                bar2.append(j-(k*j)/100)
            bar1=bar2
        return bar1
    else:
        datanew1["total_value_of_transaction_debit"]=datanew1["no_debit_card_atm_txn_value_in_mn"]+datanew1["no_debit_card_pos_txn_value_in_mn"]
        for i in lisofbanks:
            subdata1=datanew1.loc[datanew1["bank_name"]==i]
            bar1+=[subdata1["total_value_of_transaction_debit"].sum()]
        for i in range(numofiter):
            zip_object=zip(bar1,valperdec)
            bar2=[]
            for j, k in zip_object:
                bar2.append(j-(k*j)/100)
            bar1=bar2
        return bar1
```

```python
letsplot=dec_over_coming_time_val(valperdec,dorc,5,"February",dataS1,2020,["BANK OF BARODA", "PUNJAB NATIONAL BANK", "HDFC BANK LTD",'STATE BANK OF INDIA', 'ICICI BANK LTD'])
```
