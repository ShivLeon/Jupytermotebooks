```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html=urlopen("http://www.pythonscraping.com/pages/page3.html")
```

```python
bsObj=BeautifulSoup(html)
```

```python
bsObj
```

```python
for child in bsObj.find("table",{"id":"giftList"}).children:
    print(child)
```

```python
for child in bsObj.find("table",{"id":"giftList"}).descendants:
    print(child)
```

```python
bsObj1=bsObj
```

```python
bsObj1
```

```python
#Now going for sibling relation
for sibling in bsObj1.find("table",{"id":"giftList"}).tr.next_siblings:
    print(sibling)

#output of this is to print all the rows
#Except the title row(previous one had title row too)
#it happened here because objects cannot be siblings with themseleves anythime
```
